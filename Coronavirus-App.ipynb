```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_json('coronavirus_app/resources/events.json')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

```python
fig = px.scatter_geo(
    df,
    lat='lat', lon='lng', scope='asia',
    #color='size', size='size',
    hover_name='title', 
    hover_data=['title', 'event', 'link'],
    projection='natural earth', animation_frame='date', 
    title='Spread in Japan over time',
)
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
